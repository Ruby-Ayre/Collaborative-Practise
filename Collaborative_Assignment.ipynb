```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <title>Cannock Protests - Scrollytelling</title>
  <meta name="viewport" content="width=device-width, initial-scale=1" />
  <style>
    * {
      box-sizing: border-box;
    }

    body {
      margin: 0;
      font-family: system-ui, -apple-system, BlinkMacSystemFont, "Segoe UI", sans-serif;
      background: #050509;
      color: #f5f5f5;
    }

    a {
      color: #9fd5ff;
    }

    .container {
      display: flex;
      max-width: 1100px;
      margin: 0 auto;
      min-height: 100vh;
    }

    /* LEFT COLUMN – SCROLLING TEXT */
    .scrolly {
      width: 40%;
      padding: 4rem 2rem;
      position: relative;
    }

    .step {
      margin: 0 0 60vh;
      padding: 1.5rem;
      border-radius: 12px;
      background: rgba(255, 255, 255, 0.02);
      border: 1px solid rgba(255, 255, 255, 0.04);
      opacity: 0.4;
      transition: opacity 0.3s ease, transform 0.3s ease, background 0.3s ease;
    }

    .step.is-active {
      opacity: 1;
      transform: translateY(-4px);
      background: rgba(255, 255, 255, 0.06);
    }

    .step h2 {
      margin-top: 0;
      font-size: 1.35rem;
    }

    .step p {
      line-height: 1.5;
      font-size: 0.98rem;
    }

    .headline-big {
      font-size: 1.8rem;
      font-weight: 700;
      text-align: center;
      margin-bottom: 1rem;
    }

    blockquote {
      margin: 0.5rem 0;
      padding-left: 0.75rem;
      border-left: 3px solid rgba(255, 255, 255, 0.35);
      font-style: italic;
      font-size: 0.95rem;
    }

    ul {
      padding-left: 1.2rem;
      margin: 0.5rem 0 0.5rem;
      font-size: 0.95rem;
    }

    /* RIGHT COLUMN – STICKY VISUAL AREA */
    .graphic {
      width: 60%;
      position: sticky;
      top: 0;
      height: 100vh;
      display: flex;
      align-items: center;
      justify-content: center;
      padding: 2rem;
      background: radial-gradient(circle at top, #14141c 0%, #050509 55%, #000 100%);
    }

    .panel {
      position: absolute;
      max-width: 90%;
      text-align: center;
      opacity: 0;
      transform: translateY(10px);
      transition: opacity 0.4s ease, transform 0.4s ease;
      padding: 1.5rem;
      border-radius: 18px;
      /* background: rgba(0, 0, 0, 0.6); */ /* This was overriding panel-specific backgrounds */
      box-shadow: 0 18px 45px rgba(0, 0, 0, 0.5);
      backdrop-filter: blur(6px);
    }

    .panel.is-visible {
      opacity: 1;
      transform: translateY(0);
    }

    .panel h3 {
      margin-top: 0;
      font-size: 1.4rem;
    }

    .panel p {
      font-size: 0.98rem;
      line-height: 1.5;
    }

    iframe {
      width: 100%;
      min-height: 380px;
      border: none;
      border-radius: 10px;
      background: #111;
    }

    #panel-intro {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.7), rgba(0,0,0,0.8));
      background-size: cover;
      background-position: center;
    }
    h1 {
        background-color: rgba(0, 0, 0, 0.5);
        padding: 20px;
        border-radius: 8px;
    }

   /*STEP 3*/
    #panel-reasons {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.6), rgba(0,0,0,0.7));
      background-size: cover;
      background-position: center;
    }

  /*STEP 4*/
    #panel-cannock {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.6), rgba(0,0,0,0.7));
      background-size: cover;
      background-position: center;
    }

  /*STEP 5 */
    #panel-unhappy {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.6), rgba(0,0,0,0.7));
      background-size: cover;
      background-position: center;
    }
   /* STEP 6 */
    #panel-unemployment {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.6), rgba(0,0,0,0.7));
      background-size: cover;
      background-position: center;
    }

    /* STEP 7 */
    #panel-education {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.6), rgba(0,0,0,0.7));
      background-size: cover;
      background-position: center;
    }

   /* STEP 8 */
    #panel-identity {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.6), rgba(0,0,0,0.7)),
        url("https://picsum.photos/1200/800?random=7");
      background-size: cover;
      background-position: center;
    }

    /* STEP 9  */
    #panel-quotes {
      background:
        radial-gradient(circle at top, rgba(255,255,255,0.04), rgba(0,0,0,0.8));
    }

    /* STEP 10  */
    #panel-ending {
      background:
        linear-gradient(to bottom, rgba(0,0,0,0.8), rgba(0,0,0,0.9));
    }

    @media (max-width: 900px) {
      .container {
        flex-direction: column;
      }

      .scrolly,
      .graphic {
        width: 100%;
        padding: 1.5rem 1rem;
      }

      .graphic {
        position: static;
        height: auto;
      }

      .panel {
        position: static;
        opacity: 1 !important;
        transform: none !important;
        margin-bottom: 2rem;
      }

      .step {
        margin-bottom: 40vh;
      }
    }
  </style>

  <script src="https://public.flourish.studio/resources/embed.js"></script>
</head>
<body>
  <div class="container">
    <!-- LEFT COLUMN: SCROLL STEPS -->
    <div class="scrolly">
      <!-- 1. INTRO -->
      <section class="step" data-panel="intro">
        <div class="headline-big">X things you need to know about Cannock’s protests</div>
        <p>
          It’s a narrative everyone has become used to.
        </p>
        <p>
          Boats sweeping the shores, with its sailors increasing our unemployment rates, yet stealing our jobs;
          inhabiting our homes, yet squatting in hotels.
        </p>
        <p>For Cannock, these sentiments are not only felt — they’re feared.</p>
        <p><strong>But is any of it true?</strong></p>
      </section>

      <!-- 2. NATIONAL IMMIGRATION TREND -->
      <section class="step" data-panel="national">
        <h2>Is immigration actually increasing?</h2>
        <p>
          The image of small boats arriving on England’s shores is everywhere. Immigration dominates front pages
          and political speeches alike.
        </p>
        <p>
          According to GOV.UK’s net migration figures, the reality over the past decade is far more complex than a
          simple “surge”. Yet the national story is told in images of boats — not in numbers.
        </p>
      </section>

      <!-- 3. WHY IT FEELS LIKE MORE -->
      <section class="step" data-panel="reasons">
        <h2>Why does it feel like more people are arriving?</h2>
        <p>
          Even when the data doesn’t show an invasion, the sense of one can feel overwhelming. Several pressures
          collide at once:
        </p>
        <ul>
          <li>War and conflict forcing people to move</li>
          <li>A cost of living crisis and economic uncertainty</li>
          <li>Refugee law and international obligations</li>
          Each of these makes migration more visible — through headlines, footage and debate — even when the scale
          itself hasn’t exploded in the way people imagine.
        </ul>
      </section>

      <!-- 4. CANNOCK TURNS REFORM -->
      <section class="step" data-panel="cannock">
        <h2>Cannock turns to Reform</h2>
        <p>
          2025 marks the first year that Cannock has voted for Reform, a sharp turn from the 2021 district elections
          in favour of the Conservatives.
        </p>
        <p>
          Reform literally means <em>change</em>. In their manifesto, they promise quick, decisive action:
        </p>
        <ul>
          <li>“Stop the boats.”</li>
          <li>“Put British workers first.”</li>
          <li>“Take back control of our borders and our laws.”</li>
        </ul>
        <p>
          When a town is this unhappy, it’s hard to turn away from the promise of better days.
        </p>
      </section>

      <!-- 5. UNHAPPIEST PLACE -->
      <section class="step" data-panel="unhappy">
        <h2>One of the unhappiest places in the West Midlands</h2>
        <p>
          Cannock’s wellbeing scores place it among the unhappiest areas in the West Midlands, with a score of 88.8.
          It doesn’t look extreme on its own — until you place it alongside the regional average and areas like
          Wolverhampton.
        </p>
        <p>
          Discontent here doesn’t begin with immigration. It begins with how life feels.
        </p>
      </section>

      <!-- 6. UNEMPLOYMENT -->
      <section class="step" data-panel="unemployment">
        <h2>Work is scarcer here</h2>
        <p>
          Unemployment in Cannock sits above regional averages. The data shows a town where work feels less stable,
          and futures feel less secure.
        </p>
        <p>
          When jobs feel fragile, every new arrival can feel like a direct threat to your ability to stay afloat.
        </p>
      </section>

      <!-- 7. EDUCATION -->
      <section class="step" data-panel="education">
        <h2>Lower qualifications, fewer safety nets</h2>
        <p>
          Only around 23% of people in Cannock go beyond A-levels — over 10% lower than the national average. Nearly
          22% have no formal qualifications at all.
        </p>
        <p>
          But the stereotype of Cannock as simply “full of uneducated racists” doesn’t hold. Nearly half of residents
          have up to a Level 3 qualification — around 10% higher than the national average.
        </p>
        <p><strong>Cannock is not uneducated. It is misinformed.</strong></p>
      </section>

      <!-- 8. IDENTITY -->
      <section class="step" data-panel="identity">
        <h2>Who actually lives in Cannock?</h2>
        <p>
          Despite the fear of a town overrun by outsiders, the data tells a different story.
        </p>
        <p>
          As of the 2021 Census, <strong>96.5%</strong> of Cannock’s residents are UK-born, and
          <strong>97.5%</strong> identify as British.
        </p>
        <p>
          Identity is a core principle for Cannock’s residents. The fear of losing that identity is striking —
          but it is fear unmapped by data.
        </p>
      </section>

      <!-- 9. QUOTES -->
      <section class="step" data-panel="quotes">
        <h2>What protesters say</h2>
        <blockquote>“We can’t do anything about it and it’s not our choice, it’s the government’s choice.”</blockquote>
        <blockquote>“The country can’t support the people who live here, the English people and we’re just letting all the immigrants take over.”</blockquote>
        <blockquote>“Taking houses, benefits, jobs. They’re just getting full support and we have homeless people including war veterans living on the street.”</blockquote>
        <blockquote>“We’re giving benefits to those who have never contributed to our society.”</blockquote>
        <p>
          For many protesters, this isn’t a technical policy question. It’s a story of loss: if “they” gain,
          “we” must be losing.
        </p>
      </section>

      <!-- 10. ENDING -->
      <section class="step" data-panel="ending">
        <h2>When fear outruns fact</h2>
        <p>
          The personal is political. For Cannock, the threat feels real.
        </p>
        <p>
          Flag hanging is encouraged by Reform UK. Cannock’s fear feels seen; it feels protected. But it is not
          dissuaded, and it is not cured.
        </p>
        <p>
          What some see as inciting violence, others see as sharing identity. The flags mark something unmappable
          by data. The facts can show:
        </p>
        <ul>
          <li>Who lives here</li>
          <li>Who works here</li>
          <li>Who votes here</li>
        </ul>
        <p>
          But they cannot measure fear of loss, fear of invisibility, or fear of becoming powerless to a narrative
          woven through streets, news feeds and social circles.
        </p>
        <p>
          Those emotions don’t show up in spreadsheets.<br />
          They show up in protests.
        </p>
      </section>
    </div>

    <!-- RIGHT COLUMN: STICKY PANELS -->
    <div class="graphic">
      <!-- INTRO -->
      <div class="panel" id="panel-intro">
        <h3>Cannock’s anti-immigration protests</h3>
        <p>
          A town where fear of change is louder than the numbers that describe it.
        </p>
      </div>

      <!-- NATIONAL – RUBBER BOAT SEA BACKGROUND -->
      <div class="panel" id="panel-national">
        <h3>National migration trend</h3>
        <p style="font-size:0.9rem; opacity:0.85;">
          The image of small boats arriving on England’s shores is everywhere. Immigration dominates front pages
          and political speeches alike.
        </p>
        <p>
          According to GOV.UK’s net migration figures, the reality over the past decade is far more complex than a
          simple “surge”. Yet the national story is told in images of boats — not in numbers.
        </p>
      </div>

      <!-- REASONS – NEWS COLLAGE BACKGROUND -->
      <div class="panel" id="panel-reasons">
        <h3>Why it feels like more</h3>
        <p>
          Conflict, economic instability, refugee law and Brexit have all raised the visibility of migration.
          People aren’t just reacting to who moves in — they’re reacting to what they see and hear, constantly.
        </p>
      </div>

      <!-- CANNOCK TURNS REFORM -->
      <div class="panel" id="panel-cannock">
        <h3>Cannock’s political shift</h3>
        <div class="flourish-embed flourish-chart" data-src="visualisation/25982079">
          <noscript>
            <img src="https://public.flourish.studio/visualisation/25982079/thumbnail"
                 width="100%" alt="chart visualization" />
          </noscript>
        </div>
        <p style="font-size:0.8rem; opacity:0.7; margin-top:0.5rem;">
          Vote share swinging from Conservative towards Reform between 2021 and 2025.
        </p>
      </div>

      <!-- UNHAPPIEST PLACE – DATAWRAPPER -->
      <div class="panel" id="panel-unhappy">
        <h3>Happiness in Cannock</h3>
        <iframe src="https://datawrapper.dwcdn.net/gzhd4/1/" width="100%" height="500px" frameborder="0" title="Happiness comparison" aria-label="Bar Chart"></iframe>
        <p style="font-size:0.8rem; opacity:0.7; margin-top:0.5rem;">
          Cannock’s wellbeing score compared with the West Midlands average and Wolverhampton.
        </p>
      </div>

      <!-- UNEMPLOYMENT – TWO FLOURISH CHARTS -->
      <div class="panel" id="panel-unemployment">
        <h3>Unemployment in Cannock</h3>
        <div class="flourish-embed flourish-bubble-chart" data-src="visualisation/26678478">
          <noscript>
            <img src="https://public.flourish.studio/visualisation/26678478/thumbnail"
                 width="100%" alt="bubble-chart visualization" />
          </noscript>
        </div>
        <div class="flourish-embed flourish-chart" data-src="visualisation/26536677" style="margin-top:1rem;">
          <noscript>
            <img src="https://public.flourish.studio/visualisation/26536677/thumbnail"
                 width="100%" alt="chart visualization" />
          </noscript>
        </div>
        <p style="font-size:0.8rem; opacity:0.7; margin-top:0.5rem;">
          Unemployment levels and how Cannock compares to surrounding areas.
        </p>
      </div>

      <!-- EDUCATION -->
      <div class="panel" id="panel-education">
        <h3>Education in Cannock</h3>
        <div class="flourish-embed flourish-chart" data-src="visualisation/26676888">
          <noscript>
            <img src="https://public.flourish.studio/visualisation/26676888/thumbnail"
                 width="100%" alt="chart visualization" />
          </noscript>
        </div>
        <p style="font-size:0.8rem; opacity:0.7; margin-top:0.5rem;">
          Cannock’s qualification levels across no qualifications, Level 1–3 and higher education.
        </p>
      </div>

      <!-- IDENTITY -->
      <div class="panel" id="panel-identity">
        <h3>Identity in Cannock</h3>
        <div class="flourish-embed flourish-pictogram" data-src="visualisation/26497947">
          <noscript>
            <img src="https://public.flourish.studio/visualisation/26497947/thumbnail"
                 width="100%" alt="pictogram visualization" />
          </noscript>
        </div>
        <p style="font-size:0.8rem; opacity:0.7; margin-top:0.5rem;">
          The overwhelming majority of residents are UK-born and identify as British.
        </p>
      </div>

      <!-- QUOTES -->
      <div class="panel" id="panel-quotes">
        <h3>“It’s not our choice”</h3>
        <p>
          “We can’t do anything about it and it’s not our choice, it’s the government’s choice.”
        </p>
        <p style="font-size:0.9rem; opacity:0.8;">
          For protesters, decisions about who comes and who is supported feel distant and imposed.
        </p>
      </div>

      <!-- ENDING -->
      <div class="panel" id="panel-ending">
        <h3>When fear outruns fact</h3>
        <p>
          This isn’t only a protest about who crosses a border. It is a protest about who feels secure,
          who feels seen, and who believes the story of their town is being taken away from them.
        </p>
      </div >
    </div >
  </div >

  <!-- Datawrapper resizing script -->
  <script>
    window.addEventListener("message", function(event) {
      if (typeof event.data["datawrapper-height"] !== "undefined") {
        const iframes = document.querySelectorAll("iframe");
        for (const key in event.data["datawrapper-height"]) {
          for (let i = 0; i < iframes.length; i++) {
            if (iframes[i].contentWindow === event.source) {
              iframes[i].style.height = event.data["datawrapper-height"][key] + "px";
            }
          }
        }
      }
    });
  </script>

  <!-- Scroll logic: IntersectionObserver -->
  <script>
    const steps = document.querySelectorAll('.step');
    const panels = document.querySelectorAll('.panel');

    function setActive(panelId) {
      // highlight the current step
      steps.forEach(step => {
        step.classList.toggle('is-active', step.dataset.panel === panelId);
      });
      // show the matching panel
      panels.forEach(panel => {
        panel.classList.toggle('is-visible', panel.id === 'panel-' + panelId);
      });
    }

    // initial state
    setActive('intro');

    const observer = new IntersectionObserver(
      entries => {
        entries.forEach(entry => {
          if (entry.isIntersecting) {
            const panelId = entry.target.dataset.panel;
            setActive(panelId);
          }
        });
      },
      { threshold: 0.6 }
    );

    steps.forEach(step => observer.observe(step));
  </script>
</body >
</html >
```
